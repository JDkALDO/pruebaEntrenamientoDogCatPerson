## Más memoria

Los usuarios que compraron uno de los planes de pagos de Colab tienen acceso a VMs con alta capacidad de memoria cuando están disponibles. Las GPU más potentes siempre se ofrecen con las VMs con alta capacidad de memoria.
Para ver cuánta memoria te queda disponible en cualquier momento, ejecuta la siguiente celda de código. Si el resultado de la ejecución de la celda de código indica "Not using a high-RAM runtime", puedes habilitar un entorno de ejecución de RAM amplia en el menú <code>Entorno de ejecución &gt; Cambiar tipo de entorno de ejecución</code>. Luego, selecciona RAM amplia en el botón de activación del tipo de entorno de ejecución. Por último, vuelve a ejecutar la celda de código.

```
from psutil import virtual_memory
ram_gb = virtual_memory().total / 1e9
print('Your runtime has {:.1f} gigabytes of available RAM\n'.format(ram_gb))

if ram_gb < 20:
  print('Not using a high-RAM runtime')
else:
  print('You are using a high-RAM runtime!')
```
